```python
#ประกาศเพื่อเรียกใช้งาน
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot

import numpy as np

from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display
```

```python
#Lab01
sin_sig = thinkdsp.SinSignal(freq=7035, amp=35, offset=0)
sin_sig.plot()
thinkplot.config(xlabel='Time (s)')
```

```python
#Lab02
#กราฟสามเหลี่ยม
#trianglesignal(f)
#duration=signal.period*n
signal=thinkdsp.TriangleSignal(35)
duration=signal.period*5
segment=signal.make_wave(duration, framerate=10000)
segment.plot()
mpwave=signal.make_wave(duration, framerate=10000)
mpwave.apodize()
mpwave.make_audio()
```

```python
spectrum=mpwave.make_spectrum()
spectrum.plot()
```

```python
#กราฟสี่เหลี่ยม
signal=thinkdsp.SquareSignal(35)
duration=signal.period*5
segment=signal.make_wave(duration, framerate=10000)
segment.plot()
thinkplot.config(ylim=[-1.05, 1.05], legend=False)
segment.plot()
mpwave=signal.make_wave(duration, framerate=10000)
mpwave.apodize()
mpwave.make_audio()
```
